# Konsep Dasar

## Terminologi: root node, internal node, leaf node

Contoh tree:


`Root node` adalah node yang berada di paling atas / awal. Dalam kasus ini, `Root node`nya adalah pertanyaan "Fear Technology?".

`Leaf node` adalah node yang berada di bagian bawah. Contohnya adalah apple, chrome os, ubuntu linux, arch linux, dan window.

`Internal node` adalah node yang berada diantara root dan leaf node.

Ada beberapa algoritma decision tree, salah satunya adalah CART (Classification and Regression Tree).

## Gini Impurity



### Ruas Kiri:

$\begin{align}
G &= 1 - \sum\limits_{i}^n P_{i}^2 \\
&= 1 - P(biru)^2 \\
&= 1 - (\frac{4}{4})^2 = 0
\end{align}$

Nilai Gini Impurity 0 menandakan kemurnian sempurna.

### Ruas Kanan:

$\begin{align}
G &= 1 - \sum\limits_{i}^n P_{i}^2 \\
&= 1 - (P(biru)^2 + P(hijau)^2) \\
&= 1 - ((\frac{1}{6})^2 + (\frac{5}{6})^2) = 0.278
\end{align}$

### Average Gini Impurity:

$\begin{align}
G &= \frac{4}{4+6} \times 0 + \frac{6}{4+6} \times 0.278 \\
&= 0.1668
\end{align}$

## Information Gain:



# Membangun Decision Tree



$\begin{align}
G &= 1 - (P(apple)^2 + P(grape)^2 + P(lemon)^2) \\
&= 1 - ((\frac{2}{5})^2 + (\frac{2}{5})^2 + (\frac{1}{5})^2) \\
&= 0.63
\end{align}$

# Dataset

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

print(f'Dimensi Feature: {X.shape}')
print(f'Class: {set(y)}')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   test_size=0.3,
                                                   random_state=0)
```

# Classification dengan `DecisionTreeClassifier`

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=4) # maximum depth 4 layer.

model.fit(X_train, y_train)
```

# Visualisasi Model

```python
import matplotlib.pyplot as plt
from sklearn import tree

plt.rcParams['figure.dpi'] = 85 # mengatur dpi
plt.subplots(figsize=(10, 10))
tree.plot_tree(model, fontsize=10)
plt.show()
```

# Evaluasi Model

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```
