# Dataset

Referensi: https://en.wikipedia.org/wiki/Iris_flower_data_set

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

print(f'Dimensi feature: {X.shape}')
print(f'Class: {set(y)}')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   test_size=0.3, # testing size 30%
                                                   random_state=0)
```

# Classification dengan `RandomForestClassifier`

```python
from sklearn.ensemble import RandomForestClassifier # RandomForestRegressor untuk regression task.

model = RandomForestClassifier(n_estimators=100, random_state=0)

model.fit(X_train, y_train)
```

`n_estimator` = jumlah decision tree yang ingin di-train.

# Evaluasi Model

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```
