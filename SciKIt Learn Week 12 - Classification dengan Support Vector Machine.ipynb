# Konsep  Dasar

## Decision Boundary (Hyperplane)


Dalam classification task, pemisah antar class disebut `decision boundary`.

Ketika memiliki dua features, boundary-nya berupa garis. Satu feature, maka boundarynya berupa titik atau nilai threshold. Jika terdapat tiga features, boundary-nya berupa bidang datar (plane).
Empat features atau lebih, decision boundary-nya berupa hyperplane.

## Maximum Margin

Garis linear merah merupakan decision boundary yang memisahkan class biru dan hijau. Area yang diarsir kuning merupakan margin.

Margin diperoleh berdasarkan jarak terdekat antara decision boundary dengan clas yang akan dipisahkan. Setiap anggota class yang berperan dalam margin disebut support vector.

## Linearly Inseparable & Kernel Tricks

Referensi: [https://www.quora.com/What-is-the-kernel-trick]

Dengan kondisi diatas, tidak memungkinkan untuk ditarik garis linier. Oleh karena itu, SVM akan memproyeksikan data menjadi 3 dimensi sehingga kedua kelasnya dapat dipisahkan lebih mudah.

Memproyeksikan data ke dimensi yang lebih tinggi akan menambahk beban komputasi, oleh karena itu, SVM menghadirkan Kernel Tricks untuk membuat prosesnya lebih efisien.

# Dataset: The MNIST database of handwritten digits

Referensi: [https://yann.lecun.com/exdb/mnist]

```python
from sklearn.datasets import fetch_openml

X, y = fetch_openml('mnist_784', data_home='./dataset/mnist', return_X_y=True)
X.shape
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

pos = 1
for data in X[:8]: # 8 baris pertama di X
    plt.subplot(1, 8, pos)
    plt.imshow(data.reshape((28,28)),
              cmap=cm.Greys_r)
    plt.axis('off')
    pos += 1

plt.show()
```

```python
y[:8]
```

```python
# Optional
# X_train = X[:60000]
# y_train = y[:60000]
# X_test = X[60000:]
# y_test = y[60000:]

X_train = X[:1000]
y_train = y[:1000]
X_test = X[69000:]
y_test = y[69000:]
```

# Classification dengan SVC (Support Vector Classifier)

```python
from sklearn.svm import SVC

model = SVC(random_state=0)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

# Hyperparameter Tuning dengan `GridSearchCV`

```python
from sklearn.model_selection import GridSearchCV

parameters = {
    'kernel': ['rbf', 'poly', 'sigmoid'],
    'C': [0.5, 1, 10, 100],
    'gamma': ['scale', 1, 0.1, 0.01, 0.001]
}

grid_search = GridSearchCV(estimator=SVC(random_state=0),
                          param_grid= parameters,
                          n_jobs=6,
                          verbose=1,
                          scoring='accuracy')

grid_search.fit(X_train, y_train)
```

```python
print(f'Best Score: {grid_search.best_score_}') # skor akurasi terbaik.

best_params = grid_search.best_estimator_.get_params()

print(f'Best Parameters:')
for param in parameters:
    print(f'\t{param}: {best_params[param]}')
```

# Predict & Evaluate

```python
y_pred = grid_search.predict(X_test)

print(classification_report(y_test, y_pred))
```
