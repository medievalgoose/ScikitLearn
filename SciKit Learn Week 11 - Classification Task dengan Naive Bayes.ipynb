# Bayes' Theorem

### Bayes' theorem menawarkan suatu formula untuk menghitung nilai probability dari suatu event dengan memanfaatkan pengetahuan sebelumnya dari kondisi terkait; atau sering kali dikenal dengan istilah conditional probability.

### $P(A|B) = \frac{P(B|A) \times P(A)}{P(B)}$
### $P(y|X) = \frac{P(X|y) \times P(y)}{P(X)}$
### $Posterior = \frac{Likelihood \times Prior}{Evidence}$

# Pengenalan Naive Bayes Classification

## Studi Kasus 1



### MIsi: Melakukan prediksi siapa pelanggan yang melakukan pemesanan dengan diketahui pesanannya adalah lumpia dan bakso.

### $Prior Probability: P(y)$

### * $P(Asep) = 0.5$
### * $P(Joko) = 0.5$

Referensi: [https://en.wikipedia.org/wiki/Prior_probability]

### $Likelihood: P(X|y)$

### * Asep: 

### $$\begin{align*}P(lumpia, bakso|Asep) &= (0.1 \times 0.8) \\
&= 0.08 \end{align*}$$
                                   
### * Joko: 

### $$\begin{align*}P(lumpia, bakso|Joko) &= (0.3 \times 0.2) \\
&= 0.06 \end{align*}$$

## Evidence atau Normalizer: $P(X)$

### $$Evidence = \sum (Likelihood \times Prior)$$
### $$\begin{align*}P(lumpia, bakso) &= (0.08 \times 0.5) + (0.06 \times 0.5) \\
&= 0.07 \end{align*}$$

## Posterior Probability: $P(y|X)$

### * Formula:
### $$Posterior = \frac{Likelihood \times Prior}{Evidence}$$

### * Asep:
### $$\begin{align*}P(Asep|lumpia, bakso) &= \frac{0.08 \times 0.05}{0.07} \\
&= 0.57 \end{align*}$$

### * Joko:
### $$
\begin{align*}
P(Joko|lumpia, bakso) &= \frac{0.06 \times 0.05}{0.07} \\
&= 0.43
\end{align*}$$

Nilai probability Asep lebih tinggi dari Joko, maka Naive Bayes akan mengklasifikasikan Asep sebagai pembelinya.

## Studi Kasus 2



### Misi: melakukan prediksi siapa pelanggan yang melakukan pemesanan dengan diketahui pesanannya adalah siomay dan bakso.

### Evidence: $P(X)$
### $$
\begin{align*}
P(siomay, bakso) &= (0.1 \times 0.08 \times 0.5) + (0.5 \times 0.02 \times 0.5) \\
&= 0.09
\end{align*}$$

### * Asep:
### $$
\begin{align*}
P(Asep|siomay, bakso) &= \frac{(0.1 \times 0.08) \times 0.5}{0.09} \\
&= 0.444
\end{align*}$$

### * Joko:
### $$
\begin{align*}
P(Joko|siomay, bakso) 
&= \frac{(0.5 \times 0.02) \times 0.5}{0.09} \\
&= 0.555
\end{align*}$$

# Mengapa disebut Naive?

#### * Karena saat kita mendefinisikan Likelihood $P(lumpia, bakso|Asep)$, kita mengasumsikan $P(lumpia|Asep)$ conditionally independent terhadap $P(bakso|Asep)$; begitu juga sebaliknya. Sehingga dapat diformulasikan sebagai berikut:

#### $P(lumpia, bakso|Asep) = P(lumpia|Asep) \times P(bakso|Asep)$

# Dataset: Breast Cancer Wisconsin (Diagnostic)

## Load Dataset

```python
from sklearn.datasets import load_breast_cancer

print(load_breast_cancer().DESCR)
```

* Malignant - tumor ganas
* Benign - tumor jinak

```python
# load_breast_cancer? # untuk melihat documentation dari fungsi.
X, y = load_breast_cancer(return_X_y=True)
X.shape
```

## Training & Testing Set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   test_size=0.2,
                                                   random_state=0)

print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
```

## Naive Bayes dengan Scikit Learn

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
model.score(X_test, y_test)
```
